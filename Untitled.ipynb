```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import norm
%matplotlib inline
```

```python
data = pd.read_csv("IT Salary Survey EU  2020.csv")
```

```python
data.head()
```

```python
# Find column names with missing values exceeding 15%
cols_to_drop = []
for col in data.columns:
    if data[col].isnull().sum() / data.shape[0] > 0.15:
        cols_to_drop.append(col)

# Drop selected columns from dataframe
data = data.drop(cols_to_drop, axis=1)
```

```python
data.info()
```

```python
data.columns
```

```python
numeric_cols = ['Total years of experience','Years of experience in Germany','Number of vacation days','Сontract duration']  # Replace with the actual names of your columns
data[numeric_cols] = data[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
thresholds = {'Total years of experience':50, 'Yearly brutto salary (without bonus and stocks) in EUR':500000 }
for col, threshold in thresholds.items():
    data = data[data[col] <= threshold]
```

```python
skewness = stats.skew(data["Yearly brutto salary (without bonus and stocks) in EUR"])
kurtosis = stats.kurtosis(data["Yearly brutto salary (without bonus and stocks) in EUR"])

# Print the skewness and kurtosis
print("Skewness: {:.2f}".format(skewness))
print("Kurtosis: {:.2f}".format(kurtosis))
```
